```python
import pandas as pd

path="/content/drive/MyDrive/Data/car_crashes.csv"
df = pd.read_csv(path)

df.describe
```

```python
import matplotlib.pyplot as plt

#Histogram
plt.hist(df['total'], bins=20, color='skyblue')
plt.xlabel('Total Number of Crashes')
plt.ylabel('Frequency')
plt.title('Distribution of Total Crashes')
plt.show()
```

```python
#Bar chart
df['speeding'].value_counts().plot(kind='bar', color='green')
plt.xlabel('Speeding Involved (Yes/No)')
plt.ylabel('Number of Crashes')
plt.title('Crashes Involving Speeding')
plt.show()
```

```python
#Scatter plot
plt.scatter(df['alcohol'], df['total'], alpha=0.5, color='red')
plt.xlabel('Alcohol-Impaired Driving Rate')
plt.ylabel('Total Number of Crashes')
plt.title('Relationship between Alcohol-Impaired Driving and Crashes')
plt.show()
```

```python
#Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['alcohol'], vert=False)
plt.xlabel('Alcohol-Impaired Driving Rate')
plt.title('Box Plot of Alcohol-Impaired Driving Rate')
plt.show()
```

```python
#Pie chart
plt.figure(figsize=(6, 6))
counts = df['speeding'].value_counts()
labels = counts.index
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Distribution of Speeding in Crashes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
